```python
import keras
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import warnings

import pickle

from sklearn.neighbors import KernelDensity
from sklearn.ensemble import RandomForestClassifier

from tqdm import tqdm
```

# Experiment Parameters

```python
plot_step = 0.01
width = 10
n = 10000
acc_thresh = 0.8
epochs = 30

train_new = True
```

# Data Generation
## Function

```python
def generate_XOR(n):
    means = np.array([[-1], [0], [1]])
    X = np.concatenate([np.random.multivariate_normal(mean, 0.1 * np.eye(len(mean)),
                                                 size=int(n / 4)) for mean in means])

    y = np.concatenate([np.ones(int(n / 4)) * int(idx % 2) for idx in range(len(means))])

    random_indices_order = list(range(len(X)))
    np.random.shuffle(random_indices_order)
    return X[random_indices_order], np.array(y[random_indices_order]).astype('int')
```

## Training Data

```python
X, y = generate_XOR(n = n)
```

## Grid Generation

```python
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
xx = np.arange(x_min, x_max, plot_step)
```

# Construct & Train Network

```python
def get_network():
    network = keras.Sequential()
    network.add(keras.layers.Dense(width, input_shape = np.shape(X)[1:], activation = 'relu', use_bias=True, kernel_initializer = 'he_normal'))
    network.add(keras.layers.Dense(1, activation = 'sigmoid', use_bias=True, kernel_initializer = 'he_normal'))
    network.compile(loss = 'binary_crossentropy', optimizer = keras.optimizers.SGD(1e-2, momentum = 0.9), metrics = ['acc'])
    return network
```

```python
if train_new:
    acc = 0
    while acc < acc_thresh:
      network = get_network()
      network.fit(
          X, 
          y, 
          epochs = epochs, 
          verbose = True,
          batch_size = 32
      )

      acc = network.evaluate(X, y, verbose = False)[1]
      if acc >= acc_thresh:
        print("Done training. Acc = {:0.2f} > Acc Thresh = {:0.2f}.".format(acc, acc_thresh))
      else:
        print("Training New Network since Acc = {:0.2f} < Acc Thresh = {:0.2f}...".format(acc, acc_thresh))
      network.save_weights("best_weights.h5")
else:
    network = get_network()
    network.load_weights("best_weights.h5")
```

## Polytope Pyramid

```python
input_weights, ultimate_weights = [network.layers[i].get_weights()[0] for i in {0, 1}]
input_bias, ultimate_bias = [network.layers[i].get_weights()[1] for i in {0, 1}]

second_layer_preactivations = np.matmul(np.c_[xx.ravel()], input_weights) + input_bias 
second_layer_binary_activations = (second_layer_preactivations > 0).astype('int')
second_layer_activations = second_layer_preactivations * second_layer_binary_activations

ultimate_preactivations = np.matmul(second_layer_activations, ultimate_weights) + ultimate_bias
ultimate_binary_activations = (ultimate_preactivations > 0.5).astype('int')
ultimate_activations = ultimate_preactivations * ultimate_binary_activations
```

```python
partition_memberships = np.concatenate([second_layer_binary_activations, ultimate_binary_activations], axis = 1)
partition_ids = np.matmul(partition_memberships, 2 ** np.arange(0, np.shape(partition_memberships)[1]))
```

```python
figs, axs = plt.subplots(2, 3, figsize = (12 * 3, 12 * 2))
plt.subplots_adjust(hspace=0.5)
sns.set(font_scale = 2.5)
figs.text(0.5, 0.04, '', ha='center')
figs.set_facecolor('white')

for i, ax in enumerate(axs.flat):
    ax.set_facecolor('white')
    
    for side in {"right", "top"}:
        ax.spines[side].set_visible(False)
    for side in {"bottom", "left"}:
        ax.spines[side].set_visible(True)

for partition_id in partition_ids:
    idxs_of_partition = np.where(partition_ids == partition_id)[0]
    axs[0][0].plot(xx[idxs_of_partition], ultimate_preactivations[idxs_of_partition], linewidth = 6)
axs[0][0].set_ylabel("Final Pre-Activations", fontsize = 28)
axs[0][0].hlines(0, np.min(xx), np.max(xx), label = "Decision Boundary", linestyle="dashed", linewidth = 6)
axs[0][0].legend()

for partition_id in partition_ids:
    idxs_of_partition = np.where(partition_ids == partition_id)[0]
    dy = (ultimate_preactivations[idxs_of_partition[np.argmax(xx[idxs_of_partition])]] - ultimate_preactivations[idxs_of_partition[np.argmin(xx[idxs_of_partition])]])
    dx = (np.max(xx[idxs_of_partition]) - np.min(xx[idxs_of_partition]))
    axs[0][1].plot(xx[idxs_of_partition], np.ones(len(idxs_of_partition)) * dy / dx, linewidth = 6)
axs[0][1].set_ylabel("Empirical Slope", fontsize = 28)

for partition_id in partition_ids:
    idxs_of_partition = np.where(partition_ids == partition_id)[0]
    unique_layer_1_memberships_in_partition = np.where(np.unique(second_layer_binary_activations[idxs_of_partition], axis = 0)[0] == 1)[0]
    slope_constituents = []
    for layer_1_node in unique_layer_1_memberships_in_partition:
        slope_constituents.append(input_weights[0, layer_1_node]*ultimate_weights[layer_1_node, 0])
    weight_path_sum = np.sum(slope_constituents)
    
    axs[0][2].plot(xx[idxs_of_partition], np.ones(len(idxs_of_partition)) * weight_path_sum, linewidth = 6)
axs[0][2].set_ylabel("Sum of Weight-Products Along Activation Paths", fontsize = 28)

for partition_id in partition_ids:
    idxs_of_partition = np.where(partition_ids == partition_id)[0]    
    axs[1][0].plot(xx[idxs_of_partition], 1 / (1 + np.exp(-ultimate_preactivations[idxs_of_partition])), linewidth = 6)
axs[1][0].set_ylabel("Sigmoid of Final Pre-Activations", fontsize = 28)

for partition_id in partition_ids:
    idxs_of_partition = np.where(partition_ids == partition_id)[0]    
    axs[1][1].plot(xx[idxs_of_partition], network.predict(xx[idxs_of_partition]), linewidth = 6)
axs[1][1].set_ylabel("Keras Network Posterior Output", fontsize = 28)


figs.tight_layout()

plt.show()     
```
