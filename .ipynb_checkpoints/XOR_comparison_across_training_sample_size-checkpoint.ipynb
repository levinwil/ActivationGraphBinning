```python
from tensorflow import keras
import tensorflow as tf
from tensorflow.python.ops import math_ops
import numpy as np

import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['lines.linewidth'] = 6

from sklearn.neighbors import KernelDensity
from sklearn.model_selection import train_test_split

from tqdm.notebook import tqdm
```

# Experiment Parameters

```python
widths = [3, 3, 3]
num_trials = 30
```

# Construct & Train Network

```python
class Uni(keras.layers.Layer):

    def call(self, inputs):
        z_coef, z = inputs
        return z_coef / (1 + keras.backend.exp(-10*z))
    
def get_network(widths, uniform = False):
    inp = keras.Input(shape=np.shape(X)[1:])
    x = keras.layers.Dense(widths[0], activation="relu", kernel_initializer = "he_normal")(inp)
    x = keras.layers.Dense(widths[1], activation="relu", kernel_initializer = "he_normal")(x)
    x = keras.layers.Dense(widths[2], activation="relu", kernel_initializer = "he_normal")(x)
    if uniform:
        z = keras.layers.Dense(2, activation = "relu", kernel_initializer = "he_normal")(x)    
        z_coef = keras.layers.Dense(2, activation = "sigmoid", kernel_initializer = "he_normal")(x)    
        p_vals = Uni()([z_coef, z])
        out = p_vals / math_ops.reduce_sum(p_vals, axis=1, keepdims=True)
        network = keras.Model(inp, [out])
    else:
        out = keras.layers.Dense(2, activation = "softmax", kernel_initializer = "he_normal")(x)
        network = keras.Model(inp, [out])
    network.compile(loss = 'categorical_crossentropy', optimizer = keras.optimizers.Adam(3e-2), metrics = ['acc'])
    return network
```

```python
def construct_and_train_network(X, y, widths, uniform):
    network = get_network(widths, uniform)
    network.fit(
      X, 
      keras.utils.to_categorical(y), 
      epochs = 30, 
      verbose = False,
      batch_size = 32
    )

    return network

    
```

## Kernel Density Network

```python
def train_KDN(X, y):
    X_transform, X, y_transform, y = train_test_split(X, y)
    #X_transform, y_transform = X, y
    network = construct_and_train_network(X_transform, y_transform, widths, uniform = False)

    polytope_memberships = []
    last_activations = X
    
    for layer_id in range(1, len(widths) + 1):
        weights, bias = network.layers[layer_id].get_weights()
        preactivation = np.matmul(last_activations, weights) + bias
        binary_preactivation = (preactivation > 0).astype('int')
        polytope_memberships.append(binary_preactivation)
        last_activations = preactivation * binary_preactivation
        
    binary_activation = (network.predict(X) > 0.5).astype('int')
    polytope_memberships.append(binary_activation)

    polytope_memberships = np.concatenate(polytope_memberships, axis = 1)
    polytope_ids = np.matmul(polytope_memberships, 2 ** np.arange(0, np.shape(polytope_memberships)[1]))


    #an array of all of the X-values of the class-wise means in each leaf
    polytope_means_X = []

    #an array of all of the y-values (i.e. class values) of the class-wise means in each leaf
    polytope_means_y = []

    #an array of all of the number of points that comprise the means in each leaf
    polytope_means_weight = []

    #an array of all of the average variances of the points in each leaf corresponding to 
    #a single class from the class-wise mean in that leaf
    polytope_means_var = []

    polytope_KDEs = []
    for polytope_value in np.unique(polytope_ids):
        for y_val in np.unique(y):
            idxs_in_polytope_of_class = np.where((y == y_val) & (polytope_ids == polytope_value))[0]
            if len(idxs_in_polytope_of_class) > 1:
                mean_X_in_polytope_of_class = np.mean(X[idxs_in_polytope_of_class], axis = 0)

                polytope_means_X.append(mean_X_in_polytope_of_class)
                #we already know the y value, so just append it 
                polytope_means_y.append(y_val)
                #compute the number of points in that leaf corresponding to that y value
                #and append to the aggregate array
                polytope_means_weight.append(len(idxs_in_polytope_of_class))
                #compute the distances of all the points in that leaf corresponding to that y value
                #from the mean X-value of the points in that leaf corresponding to that y value
                dists_in_polytope = np.sqrt(np.sum((X[idxs_in_polytope_of_class] - mean_X_in_polytope_of_class)**2, axis = 1))
                #compute the variance as the average distance of the class-wise points in that leaf 
                #and append to the aggregate array
                polytope_means_var.append(np.mean(dists_in_polytope))
                
    #convert to numpy array so we can easily refer to multiple sets of points as ra[idxs]
    polytope_means_y, polytope_means_X, polytope_means_weight, polytope_means_var = np.array(polytope_means_y), np.array(polytope_means_X), np.array(polytope_means_weight), np.array(polytope_means_var)
    #compute the bandwidth as the average variance across the class-wise leaf means, weighted
    #by the number of points

    #an array of all of the KDEs. Each KDE will be respondsible for computing the probability 
    #that a given set of inference points belongs to that class. Thus, we have separate KDEs 
    #for each class. The KDE at index i is the KDE that is responsible for computations on 
    #y_value = i
    KDEs = []
    #loop over the y values in the leaf means 
    for y_val in np.unique(polytope_means_y):
        #gather all of the leaf means corresponding to the given y value
        polytope_means_X_of_y_val = polytope_means_X[np.where(polytope_means_y == y_val)[0]]
        #father all of the weights corresponding to the given y value
        polytope_means_weight_of_y_val = polytope_means_weight[np.where(polytope_means_y == y_val)[0]]
        #train an sklearn KDE with a gaussian kernel with the bandwidth calculated above
        polytope_means_var_of_y_val = polytope_means_var[np.where(polytope_means_y == y_val)[0]]
        bandwidth=np.average(polytope_means_var_of_y_val, weights = polytope_means_weight_of_y_val)
        kde = KernelDensity(kernel='gaussian', bandwidth=bandwidth)
        #fit the KDE on the leaf means corresponding to the given y value, weighted by
        #the weights corresponding to the given y value
        kde.fit(X = polytope_means_X_of_y_val, sample_weight = polytope_means_weight_of_y_val)

        #append the KDE to the aggregate array
        KDEs.append(kde)
        
    return KDEs
```

```python
def train_KDU(X, y):
    X_transform, X, y_transform, y = train_test_split(X, y)
    #X_transform, y_transform = X, y
    network = construct_and_train_network(X_transform, y_transform, widths, uniform = True)

    polytope_memberships = []
    last_activations = X
    
    for layer_id in range(1, len(widths) + 1):
        weights, bias = network.layers[layer_id].get_weights()
        preactivation = np.matmul(last_activations, weights) + bias
        binary_preactivation = (preactivation > 0).astype('int')
        polytope_memberships.append(binary_preactivation)
        last_activations = preactivation * binary_preactivation
        
    binary_activation = (network.predict(X) > 0.5).astype('int')
    polytope_memberships.append(binary_activation)

    polytope_memberships = np.concatenate(polytope_memberships, axis = 1)
    polytope_ids = np.matmul(polytope_memberships, 2 ** np.arange(0, np.shape(polytope_memberships)[1]))
    

    polytope_id_to_posterior = {}
    for polytope_id in np.unique(polytope_ids):
        posterior = []
        for y_val in np.unique(y):
            num_examples_of_polytope_of_class = len(np.where((polytope_ids == polytope_id) & (y == y_val))[0])
            posterior.append(num_examples_of_polytope_of_class)
        posterior = np.array(posterior) / np.sum(posterior)
        polytope_id_to_posterior[polytope_id] = posterior
        
    return network, polytope_id_to_posterior
```

```python
def generate_gaussian_parity(n):
    '''
    means = np.array([[-1, -1], [1, 1], [-1, 1], [1, -1]])
    X = np.concatenate([np.random.multivariate_normal(mean, 0.1 * np.eye(len(mean)),
                                                 size=int(n / 4)) for mean in means])

    y = np.concatenate([np.ones(int(n / 4)) * int(idx < 2) for idx in range(len(means))])
    '''
    X = np.concatenate([np.random.uniform(-1, 1, size = (n, 1)), np.random.uniform(-1, 1, size = (n, 1))], axis = 1)
    y = np.logical_xor(X[:, 0] > 0, X[:, 1] > 0).astype('int')
    
    random_indices_order = list(range(len(X)))
    np.random.shuffle(random_indices_order)
    return X[random_indices_order], np.array(y[random_indices_order]).astype('int')
```

```python
def get_brier(predicted_posterior, y):
    brier_across_y_vals = []
    for y_val in np.unique(y):
        brier_across_y_vals.append(np.nanmean((predicted_posterior[:, y_val] - (y == y_val).astype('int'))**2))
    return np.mean(brier_across_y_vals)
```

```python
X, y = generate_gaussian_parity(n = int(2e4))
X_test, y_test = generate_gaussian_parity(n = 1000)
```

```python
def get_DN_y_proba(X, y, n, X_test):
    random_indices = np.random.choice(len(X), int(n))
    X, y = X[random_indices], y[random_indices]
    network = construct_and_train_network(X, y, widths, uniform = False)
    return network.predict(X_test)
    

def get_DKN_y_proba(X, y, n, X_test):
    random_indices = np.random.choice(len(X), int(n))
    X, y = X[random_indices], y[random_indices]
    KDEs = train_KDN(X, y)
    y_proba = np.zeros((len(X_test), len(KDEs)))
    #loops over the KDEs (and thus, implicitly, the y-values)
    for y_val in range(len(KDEs)):
        #compute the (unnormalized) posterior for the KDE corresponding
        #to y_val. NOTE: we perform np.exp since the score_samples 
        #function returns the log of the probability of belonging to that 
        #class, so we must invert the log through an exponential
        y_proba[:, y_val] = np.exp(KDEs[y_val].score_samples(X_test) )

    for y_val in range(np.shape(y_proba)[1]):
        y_proba[:, y_val] /= np.sum(y_proba, axis = 1)
    
    return y_proba

def get_DKU_y_proba(X, y, n, X_test):
    random_indices = np.random.choice(len(X), int(n))
    X, y = X[random_indices], y[random_indices]
    network, polytope_id_to_posterior = train_KDU(X, y)
    
    polytope_memberships = []
    last_activations = X_test
    
    for layer_id in range(1, len(widths) + 1):
        weights, bias = network.layers[layer_id].get_weights()
        preactivation = np.matmul(last_activations, weights) + bias
        binary_preactivation = (preactivation > 0).astype('int')
        polytope_memberships.append(binary_preactivation)
        last_activations = preactivation * binary_preactivation
        
    binary_activation = (network.predict(X_test) > 0.5).astype('int')
    polytope_memberships.append(binary_activation)

    polytope_memberships = np.concatenate(polytope_memberships, axis = 1)
    polytope_ids = np.matmul(polytope_memberships, 2 ** np.arange(0, np.shape(polytope_memberships)[1]))
    
    y_proba = [polytope_id_to_posterior[polytope_id] if polytope_id in list(polytope_id_to_posterior.keys()) else [0.5, 0.5] for polytope_id in polytope_ids]
    
    return np.array(y_proba)
```

```python
def clipped_mean(ra, lower, upper):
    ra = np.array(ra)
    lower_val = np.percentile(ra, lower)
    upper_val = np.percentile(ra, upper)
    ra = ra[np.where((ra > lower_val) & (ra < upper_val))[0]]
    return np.mean(ra)
```

```python
DKN_acc_means = []
DKU_acc_means = []
network_acc_means = []

DKN_brier_means = []
DKU_brier_means = []
network_brier_means = []
n_ra = np.logspace(np.log(100) / np.log(5), np.log(50000) / np.log(5), num = 8, base = 5)
for n in tqdm(n_ra):
    DKU_y_test_proba_across_trials = [get_DKU_y_proba(X, y, n, X_test) for _ in tqdm(range(num_trials))]
    network_y_test_proba_across_trials = [get_DN_y_proba(X, y, n, X_test) for _ in tqdm(range(num_trials))]
    DKN_y_test_proba_across_trials = [get_DKN_y_proba(X, y, n, X_test) for _ in tqdm(range(num_trials))]
    

    
    DKN_accs_across_trials = []
    DKU_accs_across_trials = []
    network_accs_across_trials = []
    
    DKN_briers_across_trials = []
    DKU_briers_across_trials = []
    network_briers_across_trials = []
    for trial_idx in range(num_trials):
        DKU_accs_across_trials.append(np.nanmean(np.argmax(DKU_y_test_proba_across_trials[trial_idx], axis = 1) == y_test))
        DKN_accs_across_trials.append(np.nanmean(np.argmax(DKN_y_test_proba_across_trials[trial_idx], axis = 1) == y_test))
        network_accs_across_trials.append(np.nanmean(np.argmax(network_y_test_proba_across_trials[trial_idx], axis = 1) == y_test))
        
        DKN_briers_across_trials.append(get_brier(DKN_y_test_proba_across_trials[trial_idx], y_test))
        DKU_briers_across_trials.append(get_brier(DKU_y_test_proba_across_trials[trial_idx], y_test))
        network_briers_across_trials.append(get_brier(network_y_test_proba_across_trials[trial_idx], y_test))
        
    DKN_acc_means.append(clipped_mean(DKN_accs_across_trials, 30, 70))
    DKU_acc_means.append(clipped_mean(DKU_accs_across_trials, 30, 70))
    network_acc_means.append(clipped_mean(network_accs_across_trials, 30, 70))
        
    DKN_brier_means.append(clipped_mean(DKN_briers_across_trials, 30, 70))
    DKU_brier_means.append(clipped_mean(DKU_briers_across_trials, 30, 70))
    network_brier_means.append(clipped_mean(network_briers_across_trials, 30, 70))


    figs, ax = plt.subplots(1, 2, figsize = (18, 9))
    figs.set_facecolor("white")

    ax[0].tick_params(axis='both', which='major', labelsize=27)
    ax[0].plot(n_ra[:len(DKN_acc_means)] , DKN_acc_means, label = "Kernel Density Network", c = "red")
    ax[0].plot(n_ra[:len(DKU_acc_means)] , DKU_acc_means, label = "Uniform Network", c = "green")
    ax[0].plot(n_ra[:len(DKN_acc_means)], network_acc_means, label = "Deep Network", c = "blue")
    ax[0].hlines(1.0, 0, n_ra[len(DKN_acc_means) - 1], linestyle = "dashed", label = "Perfect Performance", color = "green")
    ax[0].hlines(0.5, 0, n_ra[len(DKN_acc_means) - 1], linestyle = "dashed", label = "Random Chance", color = "black")
    ax[0].legend(fontsize = 18)
    ax[0].set_xlabel("Number of Training Samples (logscale)", fontsize = 27)
    ax[0].set_ylabel("Test Accuracy", fontsize = 27)
    ax[0].set_xscale("log")
    ax[0].set_xticks(n_ra[:len(DKN_acc_means)].astype('int'))

    ax[1].tick_params(axis='both', which='major', labelsize=27)
    ax[1].plot(n_ra[:len(DKN_acc_means)] , DKN_brier_means, label = "Kernel Density Network", c = "red")
    ax[1].plot(n_ra[:len(DKN_acc_means)] , DKU_brier_means, label = "Uniform Network", c = "green")
    ax[1].plot(n_ra[:len(DKN_acc_means)], network_brier_means, label = "Deep Network", c = "blue")
    ax[1].hlines(0.0, 0, n_ra[len(DKN_acc_means) - 1], linestyle = "dashed", label = "Perfect Probability Forecasting", color = "green")
    ax[1].hlines(1.0, 0, n_ra[len(DKN_acc_means) - 1], linestyle = "dashed", label = "Worst Performance", color = "black")
    ax[1].legend(fontsize = 18)
    ax[1].set_xlabel("Number of Training Samples (logscale)", fontsize = 27)
    ax[1].set_ylabel("Test Mean Brier Score", fontsize = 27)
    ax[1].set_xscale("log")
    ax[1].set_xticks(n_ra[:len(DKN_acc_means)].astype('int'))
    
    figs.tight_layout()

    figs.suptitle("Posterior Estimation Comparison\nDataset: XOR with sigma = 0.1\nNetwork Widths: {}".format(widths), fontsize=27, y = 1.15)

    plt.show()
```

```python
figs, ax = plt.subplots(1, 2, figsize = (18, 9))
figs.set_facecolor("white")

ax[0].tick_params(axis='both', which='major', labelsize=27)
ax[0].plot(n_ra , DKN_acc_means, label = "Kernel Density Network", c = "red")
ax[0].plot(n_ra , DKU_acc_means, label = "Uniform Network", c = "green")
ax[0].plot(n_ra, network_acc_means, label = "Deep Network", c = "blue")
ax[0].hlines(1.0, 0, n_ra[-1], linestyle = "dashed", label = "Perfect Class Prediction", color = "green")
ax[0].hlines(0.5, 0, n_ra[-1], linestyle = "dashed", label = "Random Chance", color = "black")
ax[0].legend(fontsize = 18)
ax[0].set_xlabel("Number of Training Samples (logscale)", fontsize = 27)
ax[0].set_ylabel("Test Accuracy", fontsize = 27)
ax[0].set_xscale("log")
ax[0].set_xticks(n_ra.astype('int'))

ax[1].tick_params(axis='both', which='major', labelsize=27)
ax[1].plot(n_ra , DKN_brier_means, label = "Kernel Density Network", c = "red")
ax[1].plot(n_ra , DKU_brier_means, label = "Uniform Network", c = "green")
ax[1].plot(n_ra, network_brier_means, label = "Deep Network", c = "blue")
ax[1].hlines(0.0, 0, n_ra[-1], linestyle = "dashed", label = "Perfect Probability Forecasting", color = "green")
ax[1].hlines(0.25, 0, n_ra[-1], linestyle = "dashed", label = "Random Chance", color = "black")
ax[1].legend(fontsize = 18)
ax[1].set_xlabel("Number of Training Samples (logscale)", fontsize = 27)
ax[1].set_ylabel("Test Mean Brier Score", fontsize = 27)
ax[1].set_xscale("log")
ax[1].set_xticks(n_ra.astype('int'))


figs.tight_layout()

figs.suptitle("Posterior Estimation Comparison\nDataset: XOR\nNetwork Widths: {}".format(widths), fontsize=27, y = 1.15)

plt.show()
```
